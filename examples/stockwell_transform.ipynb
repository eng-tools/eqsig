# Convert motion to Stockwell transform and inverse


```python
import numpy as np
import eqsig
import matplotlib.pyplot as plt
```

## Load motion as array and produce Stockwell

```python
acc = np.loadtxt('test_motion_dt0p01.txt', skiprows=2)
dt = 0.01

st = eqsig.stockwell.transform(acc)

bf, sps = plt.subplots()
eqsig.stockwell.plot_tifq_vals(sps, abs(st), dt)
plt.show()
```

## Load motion as object and produce Stockwell

```python
asig = eqsig.load_asig('test_motion_dt0p01.txt')
bf, sps = plt.subplots()
eqsig.stockwell.plot_stock(sps, asig)
plt.show()
```

## Get the inverse

```python
acc_new = eqsig.stockwell.itransform(st)
plt.plot(dt * np.arange(len(acc)), acc, c='k')
# NOTE: loss of every second data point, can interpolate initial time series to avoid this
dt_new = dt * 2
plt.plot(dt_new * np.arange(len(acc_new)), acc_new, c='r')
```
